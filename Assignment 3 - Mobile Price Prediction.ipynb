Lets Clean the data Now! First let's see which features we wont need
1.Name
2.Mobile Colour(Seems irrelevant)

```python
df = raw_df.copy()
```

```python
df.head()
```

```python
#Dropping unnecessary rows
df.drop(axis = 1, columns = ['mobile_color','mobile_name'],inplace = True)
```

```python
#Removing rupee symbol from mob_price
df.mobile_price = df.mobile_price.str.replace('₹','').str.replace(',','').astype('int')
```

```python
#Label encoding dual sim feature
dual_sim_map = {'Dual Sim':1,'Single Sim':0}
df.dual_sim = df.dual_sim.map(dual_sim_map)
```

```python
#Removing all unncessary chars from disp_size feature and leaving it as cm
df.disp_size = df.disp_size.str[:5].astype('float')
```

```python
#Splitting pixels in two integer columns and making a third col with total pixels 
df.resolution = df.resolution.str[:11]
df.resolution = df.resolution.str.replace('$','').str.replace('P','').str.replace('Pi','').str.replace('i','').str.replace('*',' x ')
df['res1'] = df.resolution.str[:5].str.replace(' x','').astype('int')
df['res2'] = df.resolution.str[6:].str.replace(' ','').astype('int')
df['total_resolution'] = df.res1 * df.res2
```

```python
df.head()
```

```python
#replacing os softwares to single types
df.os = df.os.str.replace('Android Pie 9.0','Android Pie 9')
df.os = df.os.str.replace('Android Oreo 8.1.0','Android Oreo 8.1')
df.os = df.os.str.replace('Android Oreo 8.0','Android Oreo 8')
df.os = df.os.str.replace('Android 10','Android Pie 10')
df.os = df.os.str.replace('Android Nougat 7.1.2','Android Nougat 7').str.replace('Android Nougat 7.1','Android Nougat 7')
df.os = df.os.str.replace('Android Marshmallow 6.0.1','Android Marshmallow 6')
df.os = df.os.str.replace('Android Oreo 8.1','Android Oreo 8')
df.os = df.os.str.replace('Android Lollipop 5.4.1','Android Lollipop 5.1.1').str.replace('Android Lollipop 5.1.1.1','Android Lollipop 5.1.1')
```

```python
#replacing number of cores to corresponding values
core_map = {'Octa Core':8,'Quad Core':4,'Single Core':1}
df.num_cores = df.num_cores.map(core_map)
```

```python
#Correcting the value for mp_speed feature
df.mp_speed = df.mp_speed.str[:6].str.replace(' GHz','').str.replace(' GH','').str.replace(' G','').astype('float')
```

```python
#correcting value for df.ram
df.ram = df.ram.str[:2].astype('int')
```

```python
#Correcting internal memory feature
df.int_memory = df.int_memory.str[:3].str.replace(' G','').astype('int')
```

```python
df.info()
```

```python
#Correcting bluetooth
blue_map = {'Yes':1,'No':0}
df.bluetooth = df.bluetooth.map(blue_map)
```

```python
#correcting battery_power feature
df.battery_power = df.battery_power.str[:4].astype('int')
```

```python
#correcting mob width feature
df.mob_width = df.mob_width.str[:5].str.replace('mm','').str.replace('m','').astype('float')
```

```python
#correcting mob height feature
df.mob_height = df.mob_height.str[:6].str.replace('mm','').str.replace('m','').astype('float')
```

```python
#Correcting mob_depth feature
df.mob_depth = df.mob_depth.str[:3].str.replace('mm','').str.replace('m','').astype('float')
```

```python
df.mob_weight = df.mob_weight.str.replace('g','').astype('float')
```

```python
df.drop(columns = ['resolution'],inplace = True, axis = 1)
```

```python
df.select_dtypes('object')
new_df = pd.get_dummies(df,drop_first = True, columns = ['os','p_cam','f_cam','network'])
```

```python
y = new_df.mobile_price
X = new_df.drop('mobile_price',axis = 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 0)
```

```python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
```

```python
lir = LinearRegression()
lir.fit(X_train,y_train)
print(lir.score(X_train,y_train))
print(lir.score(X_test,y_test))
```

```python
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
print(rf.score(X_train,y_train))
print(rf.score(X_test,y_test))
```

```python
gnb = GaussianNB()
gnb.fit(X_train,y_train)
print(gnb.score(X_train,y_train))
print(gnb.score(X_test,y_test))
```
